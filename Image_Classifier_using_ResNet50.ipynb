```
'''
Build an image classifier using the pre-trained ResNet50 model:
 - Input will be a single jpg image
 - Output will be predicted category probabilities
1) import packages
2) load the image and size it to a standard format
3) plot the image to see what we're trying to classify and to make sure it loaded correctly
4) convert the image to a 3D Numpy array
5) the model expects a batch of images as input, so create a batch using just one image
6) define the model
7) normalize the pixels in the image to a standardized range
8) use the model to make a prediction on the image
9) print the top n predictions
'''
```

```
# 1) import packages
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```
# 2) load the image and size it to a standard format
img = image.load_img('drive/My Drive/Mittens.JPG', target_size = (224, 224))
```

```
# 3) plot the image to see what we're trying to classify and to make sure it loaded correctly
plt.imshow(img)
plt.show()
```

```
# 4) convert the image to a 3D Numpy array
img_array = image.img_to_array(img)
```

```
# 5) the model expects a batch of images as input, so create a batch using just one image
img_batch = np.expand_dims(img_array, axis = 0)
```

```
# 6) define the model
model = tf.keras.applications.resnet50.ResNet50()
```

```
# 7) normalize the pixels in the image to a standardized range
img_preprocessed = preprocess_input(img_batch)
```

```
# 8) use the model to make a prediction on the image
prediction = model.predict(img_preprocessed)
```

```
# 9) print the top n predictions
print(decode_predictions(prediction, top = 3)[0])
```
